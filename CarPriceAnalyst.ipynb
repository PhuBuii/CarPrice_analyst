```python
df['Levy'].describe()
```

```python
df['Levy'].replace(to_replace= np.nan,value=765,inplace=True)
```

```python
df['Levy']=df['Levy'].astype(float)
```

```python
df['Price']=df['Price'].astype(float)
```

```python
df['Mileage'] = df['Mileage'].str.replace('km', '').astype(int)
```

```python
df.info()
```

```python
df['Prod. year'].value_counts()
```

```python
df['Prod. year'].max()
```

```python
for dataset in [df]:
    dataset.loc[dataset['Prod. year'] <= 2000, 'Prod. year'] = 0
    dataset.loc[(dataset['Prod. year'] > 2000) & (dataset['Prod. year'] <= 2005), 'Prod. year'] = 1
    dataset.loc[(dataset['Prod. year'] > 2005) & (dataset['Prod. year'] <= 2010), 'Prod. year'] = 2
    dataset.loc[(dataset['Prod. year'] > 2010) & (dataset['Prod. year'] <= 2015), 'Prod. year'] = 3
    dataset.loc[(dataset['Prod. year'] > 2015) & (dataset['Prod. year'] <= 2020), 'Prod. year'] = 4
```

```python
df['Prod. year'].value_counts()
```

```python
# create new derived column 'Turbo' from values that have ('Turbo') in 'Engine volume'
df['Turbo'] = np.where(df['Engine volume'].str.contains('Turbo'), 'Yes', 'No')

# extract numeric part from 'Engine volume' and convert it to float type
df['Engine volume'] = df['Engine volume'].str.extract('(\d+)').astype('float64')
```

```python
df = df.rename(columns={'Mileage': 'Mileage(Km)'})
```
